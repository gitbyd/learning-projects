#### Columns that need one hot encoding
Gender, partner, PhoneService, MultipleLines, InternetService, OnlineSecurity, OnlineBackup, DeviceProtection, TechSupport, StreamingTV, StreamingMovies, Contract, PaymentMethod

```python
import sklearn
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
columns_to_encode = ['gender', 'Partner', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
                     'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 
                     'Contract', 'PaymentMethod']

encoded_data = ohe.fit_transform(df_encoded[columns_to_encode])
```

```python
# Get feature names from the OneHotEncoder
encoded_feature_names = ohe.get_feature_names_out(columns_to_encode)
encoded_feature_names
```

```python
# Convert the sparse matrix to a DataFrame
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoded_feature_names, index=df_encoded.index)
```

```python
# Concatenate with the original DataFrame (excluding the columns that were encoded)
df_final = pd.concat([df_encoded.drop(columns=columns_to_encode), encoded_df], axis=1)
df_final.head()
```

```python
df_final.dtypes
```

```python
df_final['TotalCharges'] = pd.to_numeric(df_final['TotalCharges'], errors='coerce')
df_final['Churn'] = df_final['Churn'].replace({'No': 0, 'Yes': 1}).astype(int)
```

```python
df_final.head()
```

```python
df_final.isnull().sum()
```

```python
df_final = df_final.dropna(subset=['TotalCharges'])
```

#### Splitting test and train data set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=15)
```

#### Feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train_scaled, y_train)
```

#### WE must use the same scaler object to transform the test data. And for X_test we'll use transform and not fit_transform If you were to fit a new scaler on X_test, it would learn the mean and standard deviation of your test data. This would be a form of data leakage, as you would be using information from your test set to influence a preprocessing step, which can lead to overly optimistic and misleading performance results.

```python
X_test_scaled = scaler.transform(X_test)
```

#### Predictions based on trained model lr

```python
y_pred = lr.predict(X_test_scaled)
```

```python
X_test_scaled.shape[1]
```

#### Model evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
# Calculate the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy Score: {accuracy:.2f}")
```

```python
# Calculate the accuracy score
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

#### Hyperparameter tuning

```python
# Defining the parameter grid for hyperparameter tuning
param_grid = {'C':[0.1, 1.0, 10.3],
                'solver':['liblinear', 'lbfgs']}
```

```python
from sklearn.model_selection import GridSearchCV 
```

```python
gs = GridSearchCV(LogisticRegression(), param_grid, cv=5, scoring='recall')
```

```python
gs.fit(X_train_scaled, y_train)
```

```python
gs.best_params_
```

```python
gs.best_score_
```

```python
# to check the best performing model during search
gs.best_estimator_
```

```python
y_pred_tuned = gs.best_estimator_.predict(X_test_scaled)
```

```python
confusion_matrix(y_test, y_pred_tuned)
```

```python
precision_score(y_test, y_pred_tuned)
```

```python
recall_score(y_test, y_pred_tuned)
```

### Why recal_score(y_test, y_pred_tuned) isnot same as gs.best_scroe_
1. gs.best_score_ score comes from the GridSearchCV process. It is the average recall score from the 5-fold cross-validation on the training data (X_train). It represents your model's average performance on the data it was trained on. It is an internal score that helps GridSearchCV select the best parameters.

2. recall_score is on unseen data and final, unbiased evalutaion

aka Generalization gap


We didn't see a significant improvement on the test set, but I wouldn't say hyperparameter tuning "didn't work." What it did was give us a crucial piece of information: we've likely hit the performance ceiling for a Logistic Regression model on this dataset.

The fact that the best_score_ of 55.9% was so close to our initial recall score of 54.2% tells us that simply adjusting the knobs on this model isn't going to get us much further. We've optimized what we have, and now we know its limits.

This is a valuable discovery. It tells us that our next step should not be to keep tuning the same model, but to try a more powerful algorithm.
